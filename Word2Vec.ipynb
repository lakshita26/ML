for **Word2Vec** gensim allow you to use pretained model(like google news 300) and also can train the model from scratch.

If in the dataset you see that the dataset has 75% of the words which are present in pretrained then go for it otherwise train from scratch.

Now, if we use pretrained model like google 300 then for each word of a sentence we need a vector of size 300. To resolve this problem we use **Avg Word2Vec**. 

In avg word2vec each words are avg based on the dimention(same dimention are made a single vector for the sentence by taking avg of all the words of sentence)

```python
from gensim.models import KeyedVectors
```

```python
import gensim.downloader as api

wv = api.load('word2vec-google-news-300')
# load data from google no need to download dataset
```

```python
vec_king = wv['king']
vec_king
# show the king's vector of 300 size
# (window size(hidden layer size) = 300 neurons)
```

```python
wv['man']
```

```python
wv.most_similar('man')
# show cosine similar words to man
```

```python
wv.most_similar('king')
```

```python
wv.most_similar('happy')
```

```python
wv.similarity('man', 'king')
# show how man and king are similar
```

```python
wv.similarity('html', 'programmer')
```

```python
vec = wv['king'] - wv['man'] + wv['woman']
wv.most_similar([vec])
```

```python
import os 
from os import path
from wordcloud import WordCloud
```

```python
text = """
Word2vec is a technique in natural language processing (NLP) for obtaining vector representations of words. 
These vectors capture information about the meaning of the word based on the surrounding words. 
The word2vec algorithm estimates these representations by modeling text in a large corpus. 
Once trained, such a model can detect synonymous words or suggest additional words for a partial sentence. 
Word2vec was developed by Tomáš Mikolov and colleagues at Google and published in 2013.
Word2vec represents a word as a high-dimension vector of numbers which capture relationships between words. 
In particular, words which appear in similar contexts are mapped to vectors which are nearby as measured by cosine similarity. 
This indicates the level of semantic similarity between the words, 
so for example the vectors for walk and ran are nearby, as are those for "but" and "however", and "Berlin" and "Germany".
"""
```

```python
wordcloud = WordCloud().generate(text)
wordcloud
```

```python
import matplotlib.pyplot as plt

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# lower font size
wordcloud = WordCloud(max_font_size=40).generate(text)

plt.figure()
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
