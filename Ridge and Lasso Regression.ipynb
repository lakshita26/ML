```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
df=pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
X= df['petal length (cm)'].values.reshape(-1, 1)
y= df['petal width (cm)'].values.reshape(-1, 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3)
```

# Ridge

```python
from sklearn.linear_model import Ridge
ridge=Ridge(alpha=1.0)
```

```python
ridge.fit(X_train, y_train)
```

```python
ridge.intercept_
```

```python
ridge.coef_
```

```python
pred_rid=ridge.predict([[7.5]])
pred_rid
```

```python
y_predict=ridge.predict(X_test)
```

```python
df_pred=pd.DataFrame({'Actual': y_test.squeeze(),
                      'Predicted': y_predict.squeeze()})
df_pred
```

```python
import seaborn as sns
sns.displot(y_predict - y_test, kind='kde')
```

```python
score = (ridge.score(X_test, y_test))*100
score
```

```python
from sklearn.metrics import r2_score
R2_percent=(r2_score(y_predict, y_test))*100
R2_percent
```

# Lasso

```python
from sklearn.linear_model import Lasso
lasso=Lasso(alpha=0.01)
```

```python
lasso.fit(X_train, y_train)
```

```python
lasso.intercept_
```

```python
lasso.coef_
```

```python
pred_las=lasso.predict([[7.5]])
pred_las
```

```python
y_predict_las=lasso.predict(X_test)
```

```python
df_pred_las=pd.DataFrame({'Actual': y_test.squeeze(),
                      'Predicted': y_predict_las.squeeze()})
df_pred_las
```

```python
sns.displot(y_predict_las - y_test, kind='kde')
```

```python
R2_percent_las=(r2_score(y_predict_las, y_test))*100
R2_percent_las
```
